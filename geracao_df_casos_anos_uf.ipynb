```python
import numpy as np
import pandas as pd
import pandas_profiling
import matplotlib.pyplot as plt
from collections import Counter #Contador das classes
```

```python
%matplotlib inline
```

```python
base_completa = "/home/ubuntu/Geracao_profiling/BaseCompleta_processada.csv"
df = pd.read_csv(base_completa, usecols=['UF', 'CLASSI_FIN', 'NU_ANO'])
df.shape
```

```python
header = ['ANO', 'UF', 'DENGUE', 'CHIKUNGUNYA', 'OUTRAS_DOENCAS']

dados = []

for ano in range(2013,2021):
    
    for uf in df.UF.unique():
        casos_dengue = df[(df.CLASSI_FIN == 'DENGUE') & (df.NU_ANO == ano) & (df.UF == uf)].shape[0]
        casos_chika = df[(df.CLASSI_FIN == 'CHIKUNGUNYA') & (df.NU_ANO == ano) & (df.UF == uf)].shape[0]
        casos_outros = df[(df.CLASSI_FIN == 'OUTRAS_DOENCAS') & (df.NU_ANO == ano) & (df.UF == uf)].shape[0]
        dados.append([ano, uf, casos_dengue, casos_chika, casos_outros])

df_casos = pd.DataFrame(dados, columns=header)
df_casos.head()
```

```python
df_ano = df_casos[df_casos.UF == 'PE']
ax = df_ano[['DENGUE', 'CHIKUNGUNYA', 'OUTRAS_DOENCAS']].plot(figsize=(10,5), kind='bar')
ax.set_xticklabels(df_ano["ANO"])
```
